```python
df_animes = pd.read_csv('./data/anime.csv')
df_animes['Genres_list'] = df_animes.Genres.apply(lambda x: x.split(', '))
```

```python
genres = Counter(reduce(lambda a,b: a+b, df_animes['Genres_list'], []))
print(len(genres))
g = sns.boxplot(genres)
g.set(title='Distribution of anime genre', ylabel="Count")
```

```python
g = sns.barplot(pd.DataFrame(genres.most_common(10)), x=0, y=1)
g.set(title='Count of 10 most represented genres', xlabel="Genre", ylabel="Count")
g.set_xticklabels(g.get_xticklabels(), rotation=60)
```
